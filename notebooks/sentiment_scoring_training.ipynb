```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline

model = BertForSequenceClassification.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis",num_labels=3)
tokenizer = BertTokenizer.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis")

nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

sentences = ["Operating profit rose to EUR 13.1 mn from EUR 8.7 mn in the corresponding period in 2007 representing 7.7 % of net sales.",
             "Bids or offers include at least 1,000 shares and the value of the shares must correspond to at least EUR 4,000.",
             "Raute reported a loss per share of EUR 0.86 for the first half of 2009 , against EPS of EUR 0.74 in the corresponding period of 2008.",
             ]
results = nlp(sentences)
print(results)


```

```python
import numpy as np
def calculate_sentiment_score(df):

    df['overall_score'] = df['score'] * 0.4 + df['sum_coin'] * 0.3 + df['num_comments'] * 0.3

    df['overall_score'] = df.apply(
        lambda x: x['overall_score'] * 1 if x['sentiment'] == 'positive' else
                  (x['overall_score'] * -1 if x['sentiment'] == 'negative' else x['overall_score']),
        axis=1
    )
    grouped_df = df.groupby('created_format').agg({'overall_score': 'mean'}).reset_index()
    grouped_df['sigmoid'] = grouped_df['overall_score'].apply(lambda x: 2 / (1 + np.exp(-x)) - 1)

    return grouped_df

sentiment_scores = calculate_sentiment_score(Tesla)
print(sentiment_scores)
```

```python

tesla_sentiment_scores = calculate_sentiment_score(Tesla)
nvdia_sentiment_scores = calculate_sentiment_score(nvdia)
apple_sentiment_scores = calculate_sentiment_score(Apple)
```

```python
import pandas as pd

def add_missing_dates_with_zeros(df):
    # Convert 'created_format' to datetime if it's not already
    df['created_format'] = pd.to_datetime(df['created_format'])

    # Find the min and max dates
    min_date = "2021-02-03"  #df['created_format'].min()
    max_date = "2023-12-31"   # df['created_format'].max()
    print(min_date, max_date)

    # Create a complete range of dates between min_date and max_date
    all_dates = pd.date_range(start=min_date, end=max_date, freq='D')

    # Reindex the DataFrame to include all dates, and fill missing dates with 0 score
    df_full = df.set_index('created_format').reindex(all_dates, fill_value=0)

    # Reset the index to get 'created_format' back as a column
    df_full = df_full.reset_index().rename(columns={'index': 'created_format'})

    return df_full

# Apply to each DataFrame
tesla_sentiment_scores_full = add_missing_dates_with_zeros(tesla_sentiment_scores)
nvdia_sentiment_scores_full = add_missing_dates_with_zeros(nvdia_sentiment_scores)
apple_sentiment_scores_full = add_missing_dates_with_zeros(apple_sentiment_scores)

# Show the result (optional)


```

```python
# Calculate the number of rows where 'overall_score' is 0
num_zeros = tesla_sentiment_scores_full[tesla_sentiment_scores_full['overall_score'] == 0].shape[0]

# Calculate the total number of rows
total_rows = tesla_sentiment_scores_full.shape[0]

# Calculate the percentage of 0s
percentage_zeros = (num_zeros / total_rows) * 100

# Print the result
print(f"Percentage of rows with 0 sentiment score: {percentage_zeros:.2f}%")
```
